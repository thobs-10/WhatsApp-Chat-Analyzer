```python
# for some reegular expressions 
import re
import pandas as pd
```

```python
f = open("C:/Users/Cash Crusaders/Desktop/My Portfolio/Projects/Data Science Projects/Data Science Project 9 - Whatsapp Data Analytics/Dataset/WhatsApp Chat with Transport .txt",'r',encoding='utf-8')
```

```python
# read the data
data = f.read()
```

```python
print(data)
```

```python
# apply some regular expressions
pattern = '\d{2,4}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
len(messages)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date': dates})
# convert message data type
df['message_date'] = pd.to_datetime(df['message_date'], format='%Y/%m/%d, %H:%M - ')
df.rename(columns = {'message_date':'date'}, inplace = True)
df.head()
```

```python
# seperate users and messagees
users =[]
messages = []
for mssg in df['user_message']:
    # split the user and the message
    entry = re.split('([\w\W]+?):\s', mssg)
    if entry[1:]:
        # add user name to the list
        users.append(entry[1])
        # add message to the mssges list
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
# modify the dataframe and drop the user message column
df['users'] = users
df['messages'] = messages
df.drop(columns=['user_message'], inplace = True)
df.head()
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
df.head()
```

```python
period =[]
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour)+"-"+str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
df['period'] = period
df.head()
```
